```python
# Get all classnames
classes = []
with open("/content/classes.txt") as f:
    for line in f.readlines():
        classes.append(line.split("\n")[0]) 

classes
```

```python
import shutil
import os
import random

def copy_images(parent_folder, new_subset, target_labels,datasets,test_size = 0.2,random_state = 42):
    # Get the appropriate labels
    random.seed(random_state)
    # print(f"\nUsing {dataset} labels...")
    
    # Loop through target labels
    for i in target_labels:
        # Make target directory
        for dirname in datasets:
            os.makedirs(parent_folder + "/" + new_subset + "/" + dirname + "/" + i, 
                        exist_ok=True)
        
        path = "/content/data/" + i
        # Go through labels and get appropriate classes

        for _,_,filenames in os.walk(path):
            n = len(filenames)
            
            train_len = round(int((1 - test_size) * n))
            test_len = int(n - train_len)

            random_sample = random.sample(filenames,n)
            random_sample_train = random_sample[:train_len]
            random_sample_test = random_sample[train_len:]

            # Create original image path and new path
            for name in random_sample_train:
                og_path = path + "/" + name
                new_path = parent_folder + "/" + new_subset + "/" + datasets[0] + "/" + i + "/" + name
            
                # Copy images from old path to new path
                shutil.copy2(og_path, new_path)

            for name in random_sample_test:
                og_path = path + "/" + name
                new_path = parent_folder + "/" + new_subset + "/" + datasets[1] + "/" + i + "/" + name
            
                # Copy images from old path to new path
                shutil.copy2(og_path, new_path)
```

```python
parent_folder = "/content/handImages"
new_subset = "all_classes"
datasets = ["train", "test"]

copy_images(parent_folder=parent_folder,
            new_subset=new_subset,
            target_labels=classes,
            datasets = datasets
            )
```
